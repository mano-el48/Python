```
def cria_nivel(q, nivel):
  n = []
  if q == 1:
    n.append(nivel)
    return n
  else:
    n = []
    for i in range(0,len(nivel)-1,2):
      print(f'analise do nivel - {i}')
      n.append(opernand(nivel[i],nivel[i+1]))
    q -= 1
    if q > 0:
      cria_nivel(q,n)
    arvore.append(n)
  return arvore
```
