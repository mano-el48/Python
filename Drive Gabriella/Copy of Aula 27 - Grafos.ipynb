### Grafo direcionado

Vamos começar o uso do NetworkX com a criação de um grafo direcionado que representa a rede do Twitter a seguir:

![alt text](https://docs.google.com/uc?export=download&id=18ZK4DxbunA30HGQE7F7Yhz7Tjm7tbu8n)

Com vimos em aula, trata-se de um grafo direcionado, já que as relações não são mútuas.

A criação de um grafo direcionado vazio com o NetworkX é feita por meio da função `DiGraph`. Os nós são adicionados por meio do método `add_node` que recebe como argumento o nome do nó.  Os nós criados podem ser apresentados com o uso do método `nodes`. Esse método devolve um objeto iterável que pode ser melhor visualizado com o uso da função `list`.

**Exercício:** Use as funções e métodos descritos no parágrafo anterior para criar os três nós do grafo da figura e verificar se os nós criados correspondem ao que era desejado.

```python
# Digite o seu código aqui
G = nx.DiGraph()
G.add_node('Alice')
G.add_node('Cynthia')
G.add_node('Bob')
print(G.nodes())
for node in G.nodes():
  print(node)
```

Já as arestas são criadas com o método `add_edge` que recebe como argumento os nomes dos dois nós ligados pela arestas. Como o grafo é direcionado, a ordem dos nomes indica a direção da aresta. Uma aresta criada com um nome de nó não existente cria também aquele nó. Assim, cuidado para não errar o nome dos nós ligados pela aresta. Caso haja um erro de digitação, um novo nó será criado. As arestas criadas podem ser apresentadas com o uso do método `edges`. Esse método também devolve um objetos iterável que pode ser melhor visualizado com o uso da função `list`.

```python
# Digite o seu código aqui
G.add_edge('Alice','Cynthia')
G.add_edge('Alice','Bob')
G.add_edge('Bob','Cynthia')
G.add_edge('Cynthia','Bob')
print(G.edges())
for edge in G.edges():
  print(edge)
```

Existem algumas maneiras para desenhar o grafo. Como nosso grafo tem poucos nós, o mais conveniente é usar a função do NetworkX chamada `draw_circular` que dispõe nos nós em um padrão circular. São vários os possíveis parâmetros para essa função. Porém, passaremos como argumentos apenas o grafo (parâmetro posicional) e os parâmetros nomeados `node_color`, `node_size` e `with_labels`. Para a cor, basta escolher alguma daquelas disponíveis na lista `COLORS` acima. Para o tamanho sugerimos 2000 e queremos que os nomes dos nós apareçam, então usaremos `True` para o último parâmetro.

**Exercício:** Use `draw_circular` para desenhar o grafo criado anteriormente. A função `axis` do pyplot deve ser usada com argumento `'equal'` para os eixos da área de desenho fiquem com a mesma escala.


```python
# Digite o seu código aqui
nx.draw_circular(G, node_color=COLORS[3], node_size=2000, with_labels=True)
plt.axis('equal')
```

**Exercício:** Adicione outros nós e arestas e desenhe o grafo novamente.

```python
# Digite o seu código aqui
G.add_node('Gabi')
G.add_node('Lucas')
G.add_node('Livia')
G.add_edge('Bob','Gabi')
G.add_edge('Gabi','Lucas')
G.add_edge('Lucas','Cynthia')
G.add_edge('Cynthia','Livia')
```

```python
nx.draw_circular(G, node_color=COLORS[9], node_size=2000, with_labels=True)
plt.axis('equal')
```

### Grafo não direcionado

Considere o grafo não direcionado da figura, com as distâncias entre algumas cidades dos EUA:

![alt text](https://docs.google.com/uc?export=download&id=1M7DzVxo9K7DKX-txtnlqpV2AA6WPzwMV)

Com vimos em aula, trata-se de um grafo não direcionado, já que as relações são mútuas.

A criação de um grafo não direcionado vazio com o NetworkX é feita por meio da função `Graph`. Ao invés de adicionar os nós um a um, vamos usar o método `add_nodes_from` que recebe como argumento qualquer iterável que permita percorrer os nomes desejados para os nós.

Para o nosso exemplo, usaremos o dicionário a seguir que já contém as coordenadas dos nós (embora estas não sejam usadas para a criação do grafo, já que o método itera apenas as chaves).



```python
positions = {'Chicago':(-30,30),
                 'New York City':(30,20),
                 'Los Angeles':(-60,-50),
                 'Miami':(25,-60)}
```

**Exercício:** Crie o grafo da figura (ainda sem arestas) usando `Graph` e `add_nodes_from`, e verifique se os nós foram adicionados adequadamente usando o método `nodes`.

```python
# Digite o seu código aqui
A = nx.Graph()
A.add_nodes_from(positions)
print(A.nodes())
```

A arestas também podem ser criadas a partir de qualquer iterável que contenha os pares de nomes dos nós que se relacionam. Como trata-se de um grafo não direcionado, a ordem dos nomes não importa. O método a ser usado é o `add_edges_from`.

Para o nosso exemplo, usaremos o dicionário a seguir além dos pares de nós, contém as distâncias entre as cidades (embora não sejam usadas para a criação das arestas).

```python
# Digite o seu código aqui
distances = {('Chicago','New York City'):714,
             ('Chicago','Los Angeles'):2014,
             ('Chicago','Miami'):1381,
             ('New York City','Los Angeles'):2451, 
             ('Miami','Los Angeles'):1400,
             ('New York City','Miami'):1089}
```

**Exercício:** Adicione arestas ao grafo criado no exercício anterior. Use `add_edges_from` e `distances`. Verifique se as arestas foram adicionadas adequadamente com o uso de `nodes` e de `edges`.

```python
# Digite o seu código aqui
A.add_edges_from(distances)
print(A.edges())
```

Como pare este grafo nós temos as posições desejadas para os nós, não precisamos (e nem queremos) fazer o desenho circular. A função `draw` do NetworkX tem dois parâmetros posicionais, o grafo e as posições (dicionário com nome do nó e coordenadas) e os mesmos parâmetros nomeados que `draw_circular`. No grafo anterior usamos um círculo para representar os nós. Este é o formato padrão. Podemos usar o parâmetro nomeado `node_shape` para mudar a forma do nó.

**Exercício:** Desenho o novo grafo usando a forma quadrada para o nó (`'s'`).

```python
# Digite o seu código aqui
nx.draw(A,positions, node_color=COLORS[2],node_size=2000, with_labels=True,node_shape= 's')
plt.axis('equal')
```

Você deve ter percebido no grafo desenhado na solução do exercício anterior que as etiquetas das arestas não apareceram. É necessário desenhá-las. Isso é possível com a função `draw_network_edge_labels` que recebe como parâmetros posicionais o grafo e as posições dos nós e como parâmetro nomeado `edge_labels` (um dicionário com as arestas e com os valores associados a elas). Note que isso é apenas uma apresentação gráfica e não faz do nosso grafo um grafo ponderado.

**Exercício:** Adicione os valores das arestas ao grafo do exercício anterior. É necessário redesenhar o gráfico na célula seguinte.

```python
# Digite o seu código aqui
nx.draw(A,positions, node_color=COLORS[2],node_size=2000, with_labels=True,node_shape= 's')
nx.draw_networkx_edge_labels(A,positions, edge_labels=distances)
plt.axis('equal')
```

**Exercício:** Adicione mais uma cidade e pelo menos mais uma cidade ao grafo e o redesenhe com os valores associados às arestas.

```python
# Digite o seu código aqui
A.add_node('San Diego')
A.add_edge('San Diego','Los Angeles')
positions['San Diego'] = (-55,-100)
distances[('San Diego','Los Angeles')] = 100

nx.draw(A, positions,
        node_color=COLORS[4],
        node_shape='s',
        node_size=3000,
        with_labels=True)

nx.draw_networkx_edge_labels(A,positions,edge_labels=distances)
plt.axis('equal')
```
